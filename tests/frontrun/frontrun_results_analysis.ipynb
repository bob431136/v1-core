```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  # Add this line to import numpy

# Read the CSV file
df = pd.read_csv('frontrun_results.csv')

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(df['waitBlocks'], df['profit'], alpha=0.7)
plt.xlabel('Wait Blocks')
plt.ylabel('Profit')
plt.title('Frontrun Results: Wait Blocks vs Profit')
plt.grid(True, alpha=0.3)

# Add a trend line
z = np.polyfit(df['waitBlocks'], df['profit'], 1)
p = np.poly1d(z)
plt.plot(df['waitBlocks'], p(df['waitBlocks']), "r--", alpha=0.8)

# Show the plot
plt.tight_layout()
plt.show()

# Print some basic statistics
print(df.describe())

# Find the wait block value that resulted in the maximum profit
max_profit_row = df.loc[df['profit'].idxmax()]
print(f"\nMaximum profit of {max_profit_row['profit']} was achieved with {max_profit_row['waitBlocks']} wait blocks.")
```
